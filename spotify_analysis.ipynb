```python
import os
from dotenv import load_dotenv
import spotipy
from spotipy.oauth2 import SpotifyClientCredentials
import pandas as pd



load_dotenv()

client_id = os.getenv("SPOTIFY_CLIENT_ID")
client_secret = os.getenv("SPOTIFY_CLIENT_SECRET")


client_credentials_manager = SpotifyClientCredentials(client_id=client_id, client_secret=client_secret)
sp = spotipy.Spotify(client_credentials_manager=client_credentials_manager)


artist_name = ["Alan Walker", 'Adele', 'Drake', 'Taylor Swift', 'Ed Sheeran', 'The Weeknd', 'Billie Eilish', 'Dua Lipa', 'Bruno Mars', 'Post Malone', 'Justin Bieber', 'Rihanna', 'Kendrick Lamar', 'Ariana Grande', 'BeyoncÃ©', 'Khalid', 'Camila Cabello', 'Shawn Mendes', 'Cardi B', 'Halsey']
tracks_list = []
for name in artist_name:
    result = sp.search(q='artist:' + name, type='artist', limit=1)
    artist_uri = result['artists']['items'][0]['uri']
    top_tracks = sp.artist_top_tracks(artist_uri, country='US')

    for track in top_tracks['tracks']:
        track_info = {
            'name': track['name'],
            'album': track['album']['name'],
            'release_date': track['album']['release_date'],
            'popularity': track['popularity'],
            'duration_ms': track['duration_ms'],
            'explicit': track['explicit']
        }
        tracks_list.append(track_info)


df_tracks = pd.DataFrame(tracks_list)

file_path = r'Data\top_tracks_adele.csv'
if os.path.exists(file_path):
    print("file exists in the repo")
else:
    df_tracks.to_csv('top_tracks_adele.csv', index=False)
    print("Data saved to top_tracks_adele.csv")


df_tracks
```
